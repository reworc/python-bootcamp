```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../data/names.csv")
```

```python
result = df[(df["Name"] == "Anna") & (df["Gender"] == "F") & (df["State"] == "CA")].sort_values("Year")

display(result)

plt.plot(result["Year"], result["Count"],
    color="#b06a10",
    marker = "o",
    label="Anna|CA"
    )
plt.legend()
display(plt)
```
