### Copy file names.csv from "material\Notebooks\Kursmaterialien\data.zip"

```python
# configruation: plots are displayed inline and not in separate window
%matplotlib inline                
import matplotlib.pyplot as plt

xs = []
ys = []

name = "Anna"
gender = "F"
state = "CA"

with open("names.csv", "r") as file:
    counter = 0;
    for line in file:
        data = line.strip().split(",")   # print header
        if counter == 0:
            print(data)   # print header
        
        if data[1] == name and data[3] == gender and data[4] == state :
            xs.append(int(data[2]))
            ys.append(int(data[5]))
            
        counter += 1

plt.plot(xs, ys)
plt.show()

print(xs)
print(ys)
```

```python
# configruation: plots are displayed inline and not in separate window
xs1 = []
ys1 = []

name = "Max"
gender = "M"
state = "CA"

start = 1950
end = 2000

sum = 0

with open("names.csv", "r") as file:
    counter = 0;
    for line in file:
        data = line.strip().split(",")
        counter += 1
        if counter == 1:
            print(data)   # print header            
            continue
        
        year = int(data[2])
        count = int(data[5])
        
        if year < start or year > end :
            continue
        
        if data[1] == name and data[3] == gender and data[4] == state :
            xs1.append(year)
            ys1.append(count)            
            sum += count

plt.plot(xs1, ys1)
plt.show()

print(sum)
```
